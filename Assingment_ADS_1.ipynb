```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Olympic_Data = pd.read_csv('summer.csv',  )
Country_Data = pd.read_csv('dictionary.csv',)
```

```python
print(Olympic_Data.head());
print(Country_Data.head())
```

```python
Olympics_Data_Final = Olympic_Data.merge(Country_Data, left_on='Country', right_on='Code', how='inner')
```

```python
Olympics_Data_Final
```

```python
Olympics_Data_Final = Olympics_Data_Final.drop(columns=['Country_x']);
Olympics_Data_Final = Olympics_Data_Final.rename(columns={'Country_y': 'Country'})
```

```python
Olympics_Data_Final.fillna(0,inplace=True)
```

```python
Olympics_Data_Final.head(10)
```

```python
Total_Medals = Olympics_Data_Final.groupby(['Country'])['Medal'].count()
pd.DataFrame(Total_Medals)
Total_Medals = Total_Medals.reset_index(drop=False)
Total_Medals = Total_Medals.sort_values(by = ['Medal'], ascending = False)
Total_Medals = Total_Medals.head(20)
#fig2=plt.figure(figsize=(10, 10), dpi= 80, facecolor='w', edgecolor='blue')

with plt.style.context('ggplot'):
    fig2=plt.figure(figsize=(10, 10), dpi= 80, facecolor='w', edgecolor='blue')
    barlist = plt.bar('Country','Medal',data = Total_Medals) 
    for i in range(20):
        barlist[i].set_color('green')
    barlist[0].set_color('Yellow')
    barlist[19].set_color('red')
    plt.xlabel('Country', fontsize = 16)
    plt.ylabel('Medals Won', fontsize = 16)
    plt.title('Total Medals won by Top 20 Countries', fontsize = 16)
    plt.xticks(Total_Medals['Country'], rotation=90, fontsize = 14)
    plt.show()
```

```python
Carbon_Emission_Data = pd.read_csv('List of countries by carbon dioxide emissions.csv',  )
```

```python
Carbon_Emission_Data.fillna(0,inplace=True)
```

```python
Carbon_Emission_Data
```

```python
Carbon_Emission_Data =  Carbon_Emission_Data.loc[:,['Country','Fossil CO2 emissions']]
```

```python
Carbon_Emission_Data_Country = pd.DataFrame(Carbon_Emission_Data)
```

```python
Carbon_Emission_Data_Country.reset_index(drop=False)
Carbon_Emission_Data_final = Carbon_Emission_Data_Country.sort_values(by = ['Fossil CO2 emissions'], ascending = False).head(5)
```

```python
Carbon_Emission_Data_final['Fossil CO2 emissions'] = pd.to_numeric(Carbon_Emission_Data_final['Fossil CO2 emissions'].str[:-1].astype('float64'))
Carbon_Emission_Data_final
```

```python
#ax1 = plt.subplot(121, aspect='equal')
fig3=plt.figure(figsize=(10, 10), dpi= 80, facecolor='w', edgecolor='blue')
plt.pie(Carbon_Emission_Data_final['Fossil CO2 emissions'], autopct='%1.1f%%', radius = 1,startangle=140, shadow=False, labels=Carbon_Emission_Data_final['Country'])
plt.title("Top 5 CO2 Emitting Countries For Year 2017",bbox={'facecolor':'8', 'pad':5}, y = -0.0001)
#fig3 = plt.figure(figsize=(50,20))
plt.show()
```
